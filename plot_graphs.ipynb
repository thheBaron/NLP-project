```python
import pandas as pd
import matplotlib.pyplot as plt

def load_data(file_path):
    """
    Reads the CSV file and returns a pandas DataFrame.
    """
    try:
        df = pd.read_csv(file_path)
        print(f"Successfully loaded data from {file_path}")
        return df
    except FileNotFoundError:
        print(f"Error: The file {file_path} was not found.")
        return None

def plot_scores_by_prompt_type(df, prompt_type):
    """
    Filters data by prompt_type, sorts by Final Score, and generates a bar chart
    with vertical x-axis labels and the Optimal Layer displayed on top of the bars.
    """
    # 1. Filter the data based on the prompt type
    filtered_df = df[df['Prompt Type'] == prompt_type].copy()
    
    if filtered_df.empty:
        print(f"No data found for Prompt Type: {prompt_type}")
        return

    # 2. Sort the data by 'Final Score' (descending order)
    sorted_df = filtered_df.sort_values(by='Final Score', ascending=False)

    # 3. Set up the plot
    fig, ax = plt.subplots(figsize=(14, 7)) # Adjust size as needed for readability
    
    # Create the bars
    bars = ax.bar(sorted_df['Concept'], sorted_df['Final Score'], color='steelblue', edgecolor='black')

    # 4. Format the X-axis to have vertical labels (90 degrees)
    plt.xticks(rotation=90, fontsize=10)
    
    # Set labels and title
    ax.set_xlabel('Concept', fontsize=12, fontweight='bold')
    ax.set_ylabel('Final Score', fontsize=12, fontweight='bold')
    ax.set_title(f'Final Scores by Concept (Prompt Type {prompt_type})', fontsize=14, fontweight='bold')

    # 5. Display the "Optimal Layer" on top of each bar
    for bar, layer in zip(bars, sorted_df['Optimal Layer']):
        height = bar.get_height()
        # Ensure the layer is displayed as an integer if applicable
        layer_text = str(int(layer)) if pd.notna(layer) else "N/A" 
        
        ax.text(
            bar.get_x() + bar.get_width() / 2, # X-coordinate (center of the bar)
            height,                            # Y-coordinate (top of the bar)
            layer_text,                        # The text to display (Optimal Layer)
            ha='center',                       # Horizontal alignment
            va='bottom',                       # Vertical alignment (just above the bar)
            fontsize=9,
            color='black'
        )

    # Adjust layout so the vertical labels don't get cut off
    plt.tight_layout()
    
    # Display the plot
    plt.show()

```

```python
# Replace 'experiment_results.csv' with your actual file name
file_name = 'results/BERT_run0/BERT_scores_run0.csv.csv' 

# Load the data
results_df = load_data(file_name)

if results_df is not None:
    # Generate the plot for Prompt Type 1
    print("Generating plot for Prompt Type 1...")
    plot_scores_by_prompt_type(results_df, prompt_type=1)
    
    # Generate the plot for Prompt Type 2
    # plot_scores_by_prompt_type(results_df, prompt_type=2)
```
